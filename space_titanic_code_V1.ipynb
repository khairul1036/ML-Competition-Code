```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
train_data = pd.read_csv("/content/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/content/test.csv")
test_data.head()
```

# **Work on train data**

```python
train_data = train_data.drop("Name", axis='columns') # drop single column
train_data.head()
```

```python
#duplicate value remove
print(train_data.duplicated())
train_data.drop_duplicates(inplace = True)
```

```python
#drop all null values
# df1 = train_data.dropna()
# df1.head()
```

```python
print(train_data.shape) #check shape
```

```python
train_data.info()
```

```python
#fill the null values

# x = train_data["HomePlanet"].mean() ## median()

# df["Calories"].fillna(x, inplace = True)
```

```python
# List of columns for which you want to fill NaNs with mode values ( categorical values )
columns_to_fill = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP']  # Adjust to your columns

# Fill NaNs in each column with its mode
for col in columns_to_fill:
    train_data[col].fillna(train_data[col].mode()[0], inplace=True)
```

```python
# List of columns for which you want to fill NaNs with mode values ( numerical values )
columns_to_fill = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']  # Adjust to your columns

# Fill NaNs in each column with its mode
for col in columns_to_fill:
    train_data[col].fillna(train_data[col].mean(), inplace=True)
```

```python
train_data.isnull().sum(), train_data.shape
```

```python
train_data['Destination'].unique()
```

```python
# #single column LabelEncoder
# from sklearn.preprocessing import LabelEncoder

# label_encoder = LabelEncoder()
# train_data['HomePlanet'] = label_encoder.fit_transform(train_data['HomePlanet'])
```

```python
#multi column LabelEncoder
from sklearn.preprocessing import LabelEncoder

# Define the list of columns to be label encoded
columns_to_encode = ['HomePlanet', 'Cabin', 'Destination']

# Initialize the label encoder
label_encoder = LabelEncoder()

# Apply label encoding to each specified column
for column in columns_to_encode:
    train_data[column] = label_encoder.fit_transform(train_data[column].astype(str))
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Convert binary columns to numerical
train_data['CryoSleep'] = train_data['CryoSleep'].astype(int)  # True/False to 1/0
train_data['VIP'] = train_data['VIP'].astype(int)  # True/False to 1/0
train_data['Transported'] = train_data['Transported'].astype(int)  # True/False to 1/0

# One-hot encode categorical columns
train_data = pd.get_dummies(train_data, columns=['CryoSleep','VIP'])

# PassengerId is generally not converted to numerical, but if needed, you can convert it
train_data['PassengerId'] = train_data['PassengerId'].astype(int)
```

```python
train_data.head()
```

```python
train_data['Age'].describe()
```

```python
#drop Terget column

X = train_data.drop("Transported", axis='columns') # drop single column
X.head()
```

```python
y = train_data['Transported']
y.head()
```

# **Train test split & apply algorithm models**

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=.2,random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
# Random Forest Classifier
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(X_train,y_train)
y_pred = rfc.predict(X_test)

rfc.score(X_test,y_test)
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(X_train,y_train)
y_pred = dtree.predict(X_test)

dtree.score(X_test,y_test)
```

```python
# KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=25)
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)

knn.score(X_test,y_test)
```

```python
# Support vector classification
# from sklearn.svm import SVC
# sv=SVC(kernel='linear')
# sv.fit(X_train,y_train)
# y_pred = sv.predict(X_test)

# sv.score(X_test,y_test)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
y_pred = logmodel.predict(X_test)

logmodel.score(X_test,y_test)
```

```python
# GaussianNB
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)

gnb.score(X_test,y_test)
```
